<p style="font-size: 16px; color: #FFFFFF; margin: 10px 0;">
Genetic algorithms mimic the process of natural selection to iteratively improve solutions to optimization problems. The design includes steps for generating an initial population of paths, evaluating their fitness based on total travel distance, selecting parents for crossover, applying crossover and mutation operations, and evolving the population over several generations to converge on an optimal solution.

<div style="background-color: #56644D; border-radius: 8px; padding: 20px; box-shadow: 0 2px 8px rgba(63,69,81,0.16); margin-bottom: 20px; font-family: Arial, sans-serif; color: #FFFFFF;">

<h1 style="text-align: center; font-size: 28px; font-weight: bold; margin: 0;">Step 1: Generate Initial Population with Flexible Starting Region</h1>
<p style="text-align: center; font-size: 18px; margin: 10px 0;"><strong>This function creates an initial population of random paths for the genetic algorithm.</strong></p>
<hr style="border: none; border-top: 2px solid #FFFFFF; margin: 10px 0;">
<p style="font-size: 16px; margin: 10px 0;">Each path starts at <code>start_region</code> and includes all other regions in a random order.</p>

</div>





```python
import random

def generate_initial_population(regions, start_region, population_size):
    population = []
    for _ in range(population_size):
        path = regions[:]  # Copy the list of regions
        path.remove(start_region)  # Remove start_region to allow for flexible arrangement
        random.shuffle(path)  # Randomly shuffle the other regions
        path.insert(0, start_region)  # Ensure path starts with start_region
        population.append(path)  # Add the generated path to the population
    return population
```


<div style="background-color: #56644D; border-radius: 8px; padding: 20px; box-shadow: 0 2px 8px rgba(63,69,81,0.16); margin-bottom: 20px; font-family: Arial, sans-serif; color: #FFFFFF;">

<h1 style="text-align: center; font-size: 28px; font-weight: bold; margin: 0;">Step 2: Fitness Function using Distance Matrix</h1>
<p style="text-align: center; font-size: 18px; margin: 10px 0;"><strong>This function calculates the fitness of a path based on the total travel distance.</strong></p>
<hr style="border: none; border-top: 2px solid #FFFFFF; margin: 10px 0;">
<p style="font-size: 16px; margin: 10px 0;">A shorter total distance results in a higher fitness value.</p>

</div>


```python
def fitness_function(path, distances):
    total_distance = 0
    for i in range(len(path) - 1):
        region1, region2 = path[i], path[i + 1]
        # Sum the distance between consecutive regions in the path
        total_distance += distances.get((region1, region2), distances.get((region2, region1), float('inf')))
    # Fitness is the inverse of the total distance, to prioritize shorter paths
    return 1 / total_distance if total_distance > 0 else float('inf')
```

<div style="background-color: #56644D; border-radius: 8px; padding: 20px; box-shadow: 0 2px 8px rgba(63,69,81,0.16); margin-bottom: 20px; font-family: Arial, sans-serif; color: #FFFFFF;">

<h1 style="text-align: center; font-size: 28px; font-weight: bold; margin: 0;">Step 3: Selection Function for Genetic Algorithm</h1>
<p style="text-align: center; font-size: 18px; margin: 10px 0;"><strong>This function selects two paths (parents) from the population based on their fitness.</strong></p>
<hr style="border: none; border-top: 2px solid #FFFFFF; margin: 10px 0;">
<p style="font-size: 16px; margin: 10px 0;">Paths with higher fitness have a higher probability of being selected.</p>

</div>


```python

def selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    # Calculate selection probabilities for each path based on fitness
    probabilities = [f / total_fitness for f in fitness_values]
    # Select two parents based on probabilities
    selected_index = random.choices(range(len(population)), weights=probabilities, k=2)
    return population[selected_index[0]], population[selected_index[1]]
```

%%html
<div style="background-color: #56644D; border-radius: 8px; padding: 20px; box-shadow: 0 2px 8px rgba(63,69,81,0.16); margin-bottom: 20px; font-family: Arial, sans-serif; color: #FFFFFF;">

<h1 style="text-align: center; font-size: 28px; font-weight: bold; margin: 0;">Step 4: Crossover Function with Adjustable Rate</h1>
<p style="text-align: center; font-size: 18px; margin: 10px 0;"><strong>This function creates a child path by combining parts of two parent paths.</strong></p>
<hr style="border: none; border-top: 2px solid #FFFFFF; margin: 10px 0;">
<p style="font-size: 16px; margin: 10px 0;">It maintains the start region and aims to blend paths from both parents.</p>

</div>


```python

def crossover(parent1, parent2, start_region, crossover_rate=0.8):
    if random.random() > crossover_rate:
        return parent1[:]  # No crossover; return a copy of the parent
    # Initialize child path with fixed start region
    child = [start_region] + [None] * (len(parent1) - 1)
    # Randomly select a segment from parent1 to copy into the child
    start, end = sorted(random.sample(range(1, len(parent1)), 2))
    child[start:end] = parent1[start:end]
    
    # Fill the remaining positions in the child path with regions from parent2
    for region in parent2:
        if region not in child:
            for i in range(1, len(child)):
                if child[i] is None:
                    child[i] = region
                    break
    return child
```

%%html
<div style="background-color: #56644D; border-radius: 8px; padding: 20px; box-shadow: 0 2px 8px rgba(63,69,81,0.16); margin-bottom: 20px; font-family: Arial, sans-serif; color: #FFFFFF;">

<h1 style="text-align: center; font-size: 28px; font-weight: bold; margin: 0;">Step 5: Mutation Function with Adjustable Rate</h1>
<p style="text-align: center; font-size: 18px; margin: 10px 0;"><strong>This function randomly swaps two regions in the path (excluding the start region)</strong></p>
<hr style="border: none; border-top: 2px solid #FFFFFF; margin: 10px 0;">
<p style="font-size: 16px; margin: 10px 0;">This introduces variety and helps avoid premature convergence.</p>

</div>


```python

def mutation(path, start_region, mutation_rate=0.1):
    if random.random() > mutation_rate:
        return path[:]  # No mutation; return a copy of the path
    # Ignore the start region for mutation
    path = path[1:]
    # Randomly swap two regions to create a mutation
    idx1, idx2 = random.sample(range(len(path)), 2)
    path[idx1], path[idx2] = path[idx2], path[idx1]
    return [start_region] + path  # Re-add the fixed start region
```

%%html
<div style="background-color: #56644D; border-radius: 8px; padding: 20px; box-shadow: 0 2px 8px rgba(63,69,81,0.16); margin-bottom: 20px; font-family: Arial, sans-serif; color: #FFFFFF;">

<h1 style="text-align: center; font-size: 28px; font-weight: bold; margin: 0;">Step 6: Genetic Algorithm Wrapper with Dynamic Parameters</h1>
<p style="text-align: center; font-size: 18px; margin: 10px 0;"><strong>This is the main function that runs the genetic algorithm.</strong></p>
<hr style="border: none; border-top: 2px solid #FFFFFF; margin: 10px 0;">
<p style="font-size: 16px; margin: 10px 0;">It iteratively improves the population by applying selection, crossover, and mutation.</p>

</div>


```python

def genetic_algorithm(regions, start_region, distances, population_size=10, crossover_rate=0.8, mutation_rate=0.1, generations=100):
    # Initialize the population
    population = generate_initial_population(regions, start_region, population_size)
    best_path, best_fitness = None, 0

    # Run the algorithm for a fixed number of generations
    for generation in range(generations):
        # Calculate fitness values for each path in the population
        fitness_values = [fitness_function(path, distances) for path in population]
        # Update the best path if a new best is found
        if max(fitness_values) > best_fitness:
            best_fitness = max(fitness_values)
            best_path = population[fitness_values.index(best_fitness)]

        # Generate the new population by selection, crossover, and mutation
        new_population = []
        while len(new_population) < population_size:
            parent1, parent2 = selection(population, fitness_values)
            child = crossover(parent1, parent2, start_region, crossover_rate)
            child = mutation(child, start_region, mutation_rate)
            new_population.append(child)
        population = new_population  # Replace old population with the new one

    # Return the best path and its corresponding total distance
    return best_path, 1 / best_fitness if best_fitness > 0 else float('inf')
```

<h2 style="font-size: 24px; font-weight: bold; margin: 0;">Step 1: Pheromone Evaporation Function</h2>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
Pheromones evaporate over time to simulate the decay of path desirability.
</p>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
This function reduces pheromone levels proportionally for each region connection to simulate the effects of fire spread or aftershock.
</p>

</div>



```python
def evaporate_pheromones(pheromones, evaporation_rate):
    for edge in pheromones:
        pheromones[edge] *= (1 - evaporation_rate)  # Apply evaporation factor to each edge
```

<div style="background-color: #18120F; border-radius: 8px; padding: 20px; box-shadow: 0 2px 8px rgba(63,69,81,0.16); margin-bottom: 20px; font-family: Arial, sans-serif; color: #B22222; text-align: center;">

<h2 style="font-size: 24px; font-weight: bold; margin: 0;">Step 2: Pheromone Deposition Function</h2>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
Pheromones are deposited by agents (fire trucks or police) after completing a path, reinforcing that path for future agents.
</p>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
This deposition reflects the success of rescue operations, where more successful paths are made more attractive to future agents.
</p>

</div>



```python

def deposit_pheromones(pheromones, path, contribution):
    for i in range(len(path) - 1):
        edge = (path[i], path[i + 1])
        # Increase pheromone level on each edge of the path
        pheromones[edge] = pheromones.get(edge, 0) + contribution
```

<h2 style="font-size: 24px; font-weight: bold; margin: 0;">Step 3: Calculate Edge Travel Time based on Blockage Levels</h2>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
This function calculates travel time between two regions, factoring in road blockages.
</p>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
The travel time is inversely proportional to the level of blockage (higher blockage, slower speed).
</p>

</div>


```python

def edge_travel_time(region1, region2, distances, blockages):
    avg_blockage = (blockages[region1] + blockages[region2]) / 2  # Average blockage between two regions
    # Ensure we don't divide by zero; treat max blockage as near-zero travel speed
    return distances.get((region1, region2), distances.get((region2, region1), float('inf'))) / max(1 - avg_blockage, 0.01)
```

<h2 style="font-size: 24px; font-weight: bold; margin: 0;">Step 4: Choose Next Region based on Pheromone and Distance</h2>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
This function probabilistically selects the next region based on pheromone levels (desirability) and distance.
</p>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
Alpha controls the influence of pheromones, and beta controls the influence of distance.
</p>

</div>


```python

def choose_next_region(current_region, regions, pheromones, distances, alpha=1, beta=2):
    probabilities = []  # List to hold selection probabilities for each region
    for next_region in regions:
        if next_region != current_region:
            edge = (current_region, next_region)
            tau = pheromones.get(edge, 0.1) ** alpha  # Pheromone level with alpha influence
            eta = (1 / distances.get((current_region, next_region), distances.get((next_region, current_region), float('inf')))) ** beta  # Distance influence
            probabilities.append((next_region, tau * eta))  # Add weighted probability for region

    # Handle case where all probabilities are zero
    total = sum(prob for _, prob in probabilities)
    if total == 0:
        return random.choice([region for region in regions if region != current_region])  # Random fallback

    # Normalize probabilities to sum to 1
    probabilities = [(region, prob / total) for region, prob in probabilities]
    
    # Select next region based on calculated probabilities
    return random.choices(
        [region for region, _ in probabilities],
        weights=[prob for _, prob in probabilities]
    )[0]
```

<h2 style="font-size: 24px; font-weight: bold; margin: 0;">Step 5: Simulate Agent Path</h2>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
This function simulates the path of a rescue agent across the regions.
</p>
<p style="font-size: 16px; margin: 10px 0; color: #B22222;">
It starts at the <code>start_region</code>, makes a full tour, and deposits pheromones on the chosen path.
</p>

</div>


```python

def simulate_agent_path(start_region, regions, distances, pheromones, blockages, generations=10, evaporation_rate=0.1):
    path = [start_region]  # Initialize path with starting region
    current_region = start_region  # Set current location to starting region
    for _ in range(len(regions) - 1):
        # Choose the next region based on pheromone and distance factors
        next_region = choose_next_region(current_region, regions, pheromones, distances)
        path.append(next_region)  # Add next region to the path
        current_region = next_region  # Update current region to the newly selected region
    
    # After completing the path, deposit pheromones based on path contribution
    deposit_pheromones(pheromones, path, contribution=1 / len(path))
    # Evaporate pheromones globally to simulate decay over time
    evaporate_pheromones(pheromones, evaporation_rate)
    return path  # Return the path taken by the agent
```

```python
# Minimax with Alpha-Beta Pruning for Resource Allocation in Disaster Response
# Define Max and Min Players
MAX_PLAYER, MIN_PLAYER = "Max", "Min"
# Calculate utility based on remaining resources and disaster intensities
def compute_utility(allocations, disaster_intensities):
    # Utility is the inverse of unaddressed disaster intensity
    unaddressed_disaster = sum(max(0, disaster_intensities[i] - allocations[i]) for i in range(len(disaster_intensities)))
    return 1 / (1 + unaddressed_disaster)
```

```python
# Recursive function to allocate resources along regions using Minimax with Alpha-Beta Pruning
def allocate_resources(disaster_intensities, remaining_units, depth, is_max_turn, alpha, beta):
    if depth == len(disaster_intensities) or remaining_units == 0:
        # End condition: all regions have been allocated resources, or no units are left
        allocations = [remaining_units if i == depth else 0 for i in range(len(disaster_intensities))]
        return compute_utility(allocations, disaster_intensities), allocations

    # For each possible allocation from current region to the next
    if is_max_turn:
        max_utility, best_allocation = -math.inf, []
        for units in range(remaining_units + 1):
            utility, allocation = allocate_resources(
                disaster_intensities, remaining_units - units, depth + 1, False, alpha, beta
            )
            if utility > max_utility:
                max_utility, best_allocation = utility, allocation
            alpha = max(alpha, utility)
            if beta <= alpha:
                break  # Alpha-beta pruning
        return max_utility, [remaining_units - sum(best_allocation[:depth])] + best_allocation

    else:  # Min turn
        min_utility, best_allocation = math.inf, []
        for units in range(remaining_units + 1):
            utility, allocation = allocate_resources(
                disaster_intensities, remaining_units - units, depth + 1, True, alpha, beta
            )
            if utility < min_utility:
                min_utility, best_allocation = utility, allocation
            beta = min(beta, utility)
            if beta <= alpha:
                break  # Alpha-beta pruning
        return min_utility, [remaining_units - sum(best_allocation[:depth])] + best_allocation
```

```python
# High-level function to start the allocation using Minimax
def resource_allocation_minimax(total_units, disaster_intensities):
    # Initialize alpha and beta for alpha-beta pruning
    alpha, beta = -math.inf, math.inf
    max_utility, allocation = allocate_resources(disaster_intensities, total_units, 0, True, alpha, beta)
    return allocation, max_utility
```

```python
# Define disaster intensities in regions and total units available
disaster_intensities = [5, 8, 3, 7, 6]  # Example disaster intensities
total_units = 10  # Example total units available

# Run resource allocation using Minimax
allocation, utility = resource_allocation_minimax(total_units, disaster_intensities)

# Output the results
allocation, utility
```

The goal of Task 5 is to model uncertainties in the rescue operations for the earthquake-affected regions using a Bayesian Network. <h2 style="font-size: 32px; font-weight: bold; color: #3B2320; margin-top: 25px;">Network Structure</h2>
<p style="font-size: 18px; font-weight: 500; color: #3B2320; margin-top: 15px;">
We designed the Bayesian Network to represent dependencies between factors influencing the success of rescue operations in each region. For this simulation, each region has three primary nodes:
</p>
<ul style="font-size: 18px; font-weight: 500; color: #3B2320; text-align: left; padding-left: 20px;">
    <li><strong>Blocked Roads</strong> - Representing the probability of road blockages that may hinder rescue operations.</li>
    <li><strong>Fire Spread</strong> - Representing the probability of fire spread, affecting the safety and accessibility of a region.</li>
    <li><strong>Success Rate</strong> - The main outcome node, indicating the likelihood of a successful rescue operation, dependent on the conditions of both road blockage and fire spread.</li>
</ul>
<p style="font-size: 18px; font-weight: 500; color: #3B2320;">
The network structure for each region (e.g., R1) is defined as:
</p>
<pre style="background-color: #3B2320; color: #FFFFFF; padding: 10px; border-radius: 5px; font-size: 18px; font-weight: bold;">
BlockedRoads_R1 -> SuccessRate_R1
FireSpread_R1 -> SuccessRate_R1
</pre>
<p style="font-size: 18px; font-weight: 500; color: #3B2320;">
This structure ensures that SuccessRate is conditioned on both Blocked Roads and Fire Spread, allowing us to infer how these factors jointly influence the outcome.
</p>

<h2 style="font-size: 32px; font-weight: bold; color: #3B2320; margin-top: 25px;">Conditional Probability Tables (CPTs)</h2>
<p style="font-size: 18px; font-weight: 500; color: #3B2320; margin-top: 15px;">
For each node, we define a Conditional Probability Table (CPT):
</p>
<ul style="font-size: 18px; font-weight: 500; color: #3B2320; text-align: left; padding-left: 20px;">
    <li><strong>Blocked Roads</strong> and <strong>Fire Spread</strong> nodes are defined with basic probabilities that reflect the likelihood of road blockage or fire presence based on the specifications. For instance, R1 has a high blockage rate, so BlockedRoads_R1 has a higher probability of being "1" (blocked).</li>
    <li><strong>Success Rate</strong> is defined with a CPT that captures the joint influence of Blocked Roads and Fire Spread. This table provides probabilities for successful and unsuccessful outcomes based on combinations of the conditions of these two nodes.</li>
</ul>
<p style="font-size: 18px; font-weight: 500; color: #3B2320;">
For example, high blockage and high fire spread lead to a lower success probability.
</p>

<h2 style="font-size: 32px; font-weight: bold; color: #3B2320; margin-top: 25px;">Implementation</h2>
<p style="font-size: 18px; font-weight: 500; color: #3B2320; margin-top: 15px;">
Using the <code>pgmpy</code> library:
</p>
<ul style="font-size: 18px; font-weight: 500; color: #3B2320; text-align: left; padding-left: 20px;">
    <li>We defined the Bayesian Network structure by specifying the dependencies between nodes.</li>
    <li>We created and added CPTs for each node, particularly focusing on accurately modeling the conditional dependencies for the SuccessRate node.</li>
    <li>We verified the model’s correctness using <code>model.check_model()</code> to ensure all dependencies and probabilities were correctly implemented.</li>
</ul>

<h2 style="font-size: 32px; font-weight: bold; color: #3B2320; margin-top: 25px;">Inference</h2>
<p style="font-size: 18px; font-weight: 500; color: #3B2320; margin-top: 15px;">
To utilize the Bayesian Network in decision-making, we performed inference using <code>VariableElimination</code>:
</p>
<ul style="font-size: 18px; font-weight: 500; color: #3B2320; text-align: left; padding-left: 20px;">
    <li>We queried the network to calculate the SuccessRate given specific conditions on BlockedRoads and FireSpread (e.g., <code>BlockedRoads_R1 = 1</code> and <code>FireSpread_R1 = 0</code>).</li>
    <li>This inference allows rescue agents to assess the likelihood of successful operations based on observed conditions, enabling them to make informed decisions under uncertainty.</li>
</ul>

</div>


```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

# Define the structure of the Bayesian Network
# Each region has nodes for Blocked Roads and Fire Spread affecting Success Rate

model = BayesianNetwork([
    ('BlockedRoads_R1', 'SuccessRate_R1'), 
    ('FireSpread_R1', 'SuccessRate_R1'),
])

# Initial damage data for each region
damage_data = {
    'R1': {'roads_blocked': 0.60, 'fire_spread': 0.35},  # High damage
    'R2': {'roads_blocked': 0.40, 'fire_spread': 0.25},  # Medium damage
    'R3': {'roads_blocked': 0.15, 'fire_spread': 0.05},  # Medium damage
    'R4': {'roads_blocked': 0.35, 'fire_spread': 0.30},  # High damage
    'R5': {'roads_blocked': 0.05, 'fire_spread': 0.03}   # Low damage
}

# Dynamic CPT generation for Blocked Roads and Fire Spread based on initial data
def create_dynamic_cpd(variable, probability):
    """Creates a TabularCPD for a binary variable with a specified probability."""
    return TabularCPD(variable=variable, variable_card=2, values=[[1 - probability], [probability]])

# Generate CPDs based on damage data for R1 as an example
cpd_blocked_roads_r1 = create_dynamic_cpd('BlockedRoads_R1', damage_data['R1']['roads_blocked'])
cpd_fire_spread_r1 = create_dynamic_cpd('FireSpread_R1', damage_data['R1']['fire_spread'])

# Define CPT for SuccessRate in R1, based on Blocked Roads and Fire Spread
cpd_success_rate_r1 = TabularCPD(
    variable='SuccessRate_R1', variable_card=2,
    # Higher probability of failure when either roads are blocked or fires are spreading
    values=[
        [0.7, 0.5, 0.4, 0.1],  # Success Probability (adjusted for realistic probabilities)
        [0.3, 0.5, 0.6, 0.9]   # Failure Probability
    ],
    evidence=['BlockedRoads_R1', 'FireSpread_R1'],
    evidence_card=[2, 2]
)

# Add CPDs to the model
model.add_cpds(cpd_blocked_roads_r1, cpd_fire_spread_r1, cpd_success_rate_r1)

# Check if the model is correctly defined
assert model.check_model()

# Perform inference for a particular region's Success Rate given observed conditions
inference = VariableElimination(model)
```
